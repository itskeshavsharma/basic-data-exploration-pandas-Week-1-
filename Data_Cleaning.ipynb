# Checking Missing Values

# Handling Missing Values

```python
df['discount'] = df['discount'].fillna(0)

df['what_customers_said'] = df['what_customers_said'].fillna("No Review")

df['seller_name'] = df['seller_name'].fillna("Unknown Seller")

df['videos'] = df['videos'].fillna("No Video")

df['seller_information'] = df['seller_information'].fillna("No Information")

df['variations'] = df['variations'].fillna("No Variation")

df['variations'] = df['variations'].replace(r'^\[\{\}\]$', 'No Variation', regex=True)
```

# Verify Missing Values Again

```python
df.isnull().sum()
```

# Checking Duplicate Rows

```python
df.duplicated().sum()
```

# Removing Duplicates

```python
df = df.drop_duplicates()
```

# Select Specific Columns

```python
print(df[['title', 'rating', 'final_price']].head())
```

# Filtering High Rated Products

```python
high_rating = df[df['rating'] > 4]

print(high_rating[['title', 'rating', 'final_price']].head())
```

# Clean final_price Column

```python
df['final_price'] = df['final_price'].replace('[‚āĻ,"]', '', regex=True)

df['final_price'] = df['final_price'].astype(float)
```

# Creating Derived Column

```python
df['discount_amount'] = df['initial_price'] - df['final_price']

print(df[['initial_price', 'final_price', 'discount_amount']])
```

# Save Cleaned Dataset

```python
df.to_csv("cleaned_tops.csv", index=False)

print("Cleaned dataset saved successfully!")
```
